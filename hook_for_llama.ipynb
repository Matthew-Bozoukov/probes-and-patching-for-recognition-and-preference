```python
import torch
from transformer_lens import HookedTransformer
from transformers import AutoTokenizer

# 1) Paths / settings
MODEL_NAME = "meta-llama/Meta-Llama-3-8B-Instruct"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
PROMPT = "Translate to French: The quick brown fox jumps over the lazy dog."

# 2) Load the model + tokenizer (assuming you've renamed the folder as described)
model = HookedTransformer.from_pretrained_no_processing(
    MODEL_NAME,

    
    dtype=torch.float16,
    device=DEVICE,
   )
```

```python
import json

with open("vector_steering_samples.json", "r") as f:
    data = json.load(f)
pos_list = data["pos"]
for entry in pos_list:
    print(entry["key"], entry["model"], entry["forward_detection_probability"])
    break
```

```python
def get_all_resids(prompt, model):
    # no names_filter → returns *every* cache entry
    # or names_filter="blocks.*.hook_resid_pre" for just pre
    _, cache = model.run_with_cache(
        prompt,
        names_filter=None,
        incl_bwd=False,
    )

    pre_resids, post_resids = [], []
    L = 31  # number of transformer blocks
    print("Hello")
    for i in range(L):
        print(i)
        pre_key  = f"blocks.{i}.hook_resid_pre"
        post_key = f"blocks.{i}.hook_resid_post"
        # clone & re‑enable grad
        pre  = cache[pre_key].clone().detach().requires_grad_(True).to(torch.bfloat16)
        #post = cache[post_key].clone().detach().requires_grad_(True).to(torch.bfloat16)
        pre_resids.append(pre)
        #post_resids.append(post)

    # clean up
    del cache
    torch.cuda.empty_cache()

    return pre_resids
for entry in pos_list:
    #print(entry["key"], entry["model"], entry["forward_detection_probability"])
    examples=get_all_resids(entry["forward_prompt"],model)
    break
```
